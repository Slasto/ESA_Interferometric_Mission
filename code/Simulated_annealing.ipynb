```python
from problems.golomb_simple import udp
from scipy.optimize import dual_annealing
from modules.plotter import print_result

fitness = lambda solution: udp.fitness(solution)[0]
```

## Dual Simulated annealing
docs : https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.dual_annealing.html

```python
bounds = [(-1.0, 1.0) for _ in range(30)]

best = dual_annealing(
    fitness,
    bounds,
    args=(),
    maxiter=5000,
    minimizer_kwargs=None,
    initial_temp=5230.0,
    restart_temp_ratio=2e-05,
    visit=2.62,
    accept=-5.0,
    maxfun=10000000.0,
    seed=None,
    no_local_search=False, # is set to True, a traditional Simulated Annealing will be performed with no local search strategy applied. 
    callback=None,
    x0=None,
)

print_result(udp, best.x,image_path="../data/nebula.jpg")
```
