## Dual Simulated annealing
docs : https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.dual_annealing.html

```python
from modules.golomb_problem import init_simple_problem, init_medium_problem
from modules.plotter import print_result
from scipy.optimize import dual_annealing

from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('jpeg')  # Configures Matplotlib to output inline plots as JPEGs, reducing the file size.
```

## Simple problem

### Stock

```python
UDP = init_simple_problem()
BOUNDS = [(-1.0, 1.0) for _ in range(UDP.n_sat*6)]

result = dual_annealing(
    lambda x: UDP.fitness(x)[0],
    BOUNDS,
    maxiter=5000,
    minimizer_kwargs=None,
    initial_temp=5230.0,
    restart_temp_ratio=2e-05,
    visit=2.62,
    accept=-5.0,
    maxfun=10000000.0,
    seed=None,
    no_local_search=False,  # is True, a traditional Simulated Annealing will be performed with no local search strategy applied.
    callback=None,
    x0=None
)

print_result(UDP, result.x, show_simulated_reconstruction=True)
del UDP, BOUNDS ,result
```

### Force Golomb

```python
UDP = init_simple_problem()
BOUNDS = [(-1.0, 1.0) for _ in range(UDP.n_sat*6)]

result = dual_annealing(
    lambda x: UDP.fitness_impl(x, reduce_fill_if_not_optimal=True)[0],
    BOUNDS,
    args=(),
    maxiter=7500,
    minimizer_kwargs=None,
    initial_temp=6230.0,
    restart_temp_ratio=2e-05,
    visit=2.62,
    accept=-5.0,
    maxfun=10000000.0,
    seed=None,
    no_local_search=False,  # is True, a traditional Simulated Annealing will be performed with no local search strategy applied.
    callback=None,
    x0=None,
)

print_result(UDP, result.x, show_simulated_reconstruction=True)
del UDP, BOUNDS, result
```
